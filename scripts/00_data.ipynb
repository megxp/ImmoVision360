## 1. Chargement des Données

```python
import pandas as pd

# Chargement du fichier 'listings.csv' depuis Google Drive
drive_listings_df = pd.read_csv('/content/drive/MyDrive/[Cours] Collecte&ExplorationData/listings.csv')

# Chargement du fichier 'reviews.csv' depuis Google Drive
drive_reviews_df = pd.read_csv('/content/drive/MyDrive/[Cours] Collecte&ExplorationData/reviews.csv')
```

## 2. Exploration des Données 'listings'

### 2.1 Aperçu Général du DataFrame `drive_listings_df`

```python
print('\nPremières 5 lignes :')
display(drive_listings_df.head())
print('\nColonnes :')
print(drive_listings_df.columns)
print('\nTaille (lignes, colonnes) :')
print(drive_listings_df.shape)
```

### 2.2 Examen d'un Enregistrement Unique en format paragraphe

```python
# Sélection du premier enregistrement
record = drive_listings_df.iloc[0]

# Formatage de l'enregistrement en paragraphe
formatted_record = []
for column, value in record.items():
    formatted_record.append(f"{column}: {value}")

# Joindre toutes les parties formatées avec un séparateur pour former un paragraphe
paragraph_output = "\n ".join(formatted_record)

print(paragraph_output)
```

## 3. Exploration des Données 'reviews'

### 3.1 Aperçu Général du DataFrame `drive_reviews_df`

```python
print('\nPremières 5 lignes :')
display(drive_reviews_df.head())
print('\nColonnes :')
print(drive_reviews_df.columns)
print('\nTaille (lignes, colonnes) :')
print(drive_reviews_df.shape)
```

### 3.2 Commentaires pour une Annonce Spécifique

```python
# Définir l'ID de l'annonce à rechercher
listing_id_to_search = 2719440

# Filtrer le DataFrame drive_reviews_df pour l'ID d'annonce spécifié
comments_for_listing = drive_reviews_df[drive_reviews_df['listing_id'] == listing_id_to_search]

# Afficher les commentaires pour cette annonce
if not comments_for_listing.empty:
    print(f"Commentaires pour l'annonce {listing_id_to_search}:")
    for index, row in comments_for_listing.iterrows():
        print(f"- {row['comments']}")
else:
    print(f"Aucun commentaire trouvé pour l'annonce {listing_id_to_search}.")
```

## 4. Analyse des Quartiers

### 4.1 Liste des Quartiers Disponibles (`neighbourhood_cleansed`)

```python
# Obtenir toutes les valeurs uniques de la colonne 'neighbourhood_cleansed'
unique_neighbourhoods = drive_listings_df['neighbourhood_cleansed'].unique()

# Afficher la liste des quartiers uniques
print("Liste des 'neighbourhood_cleansed' uniques:")
for neighbourhood in unique_neighbourhoods:
    print(f"- {neighbourhood}")
```

### 4.2 Comptage des Annonces dans un Quartier Spécifique (Exemple : 'Élysée')

Note : L'exécution précédente de ce code a renvoyé 0, car 'Élysée' n'est pas une valeur exacte dans la colonne `neighbourhood_cleansed`. La liste ci-dessus montre les noms exacts des quartiers disponibles.

```python
# Filtrer le DataFrame pour les annonces dans 'Le Élysée' en utilisant 'neighbourhood_cleansed'
e_listings = drive_listings_df[drive_listings_df['neighbourhood_cleansed'] == 'Élysée']

# Compter le nombre d'annonces
num_e_listings = len(e_listings)

print(f"Nombre d'éléments situés dans le Marais: {num_e_listings}")
```
